```python
def soma(matriz1, matriz2):
    n = len(matriz1)
    resultado = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            resultado[i][j] = matriz1[i][j] + matriz2[i][j]
    
    return resultado


matriz1 = [[1, 2], [3, 4]]
matriz2 = [[5, 6], [7, 8]]

resultado = soma(matriz1, matriz2)
print(resultado)
```

```python
def subtracao(matriz1, matriz2):
    n = len(matriz1)
    resultado = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            resultado[i][j] = matriz1[i][j] - matriz2[i][j]
    
    return resultado

matriz1 = [[1, 2], [3, 4]]
matriz2 = [[5, 6], [7, 8]]

resultado = subtracao(matriz1, matriz2)
print(resultado)
```

```python
def transpostaMatriz(matriz):
    n = len(matriz)
    transposta = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            transposta[j][i] = matriz[i][j]
    
    return transposta

matriz = [[1, 2], [3, 4]]

transposta = transpostaMatriz(matriz)
print(transposta)
```

```python
def multiplicacao(matriz1, matriz2):
    n = len(matriz1)
    resultado = [[0 for _ in range(n)] for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                resultado[i][j] += matriz1[i][k] * matriz2[k][j]
    
    return resultado


matriz1 = [[1, 2], [3, 4]]
matriz2 = [[5, 6], [7, 8]]

resultado = multiplicacao(matriz1, matriz2)
print(resultado)
```

```python
def criar_tabuleiro():
    tabuleiro = [[' ' for _ in range(3)] for _ in range(3)]
    return tabuleiro

def imprimir_tabuleiro(tabuleiro):
    for linha in tabuleiro:
        print('|'.join(linha))
        print('-' * 5)

def verificar_vitoria(tabuleiro, jogador):
    # Verificar linhas
    for linha in tabuleiro:
        if linha.count(jogador) == 3:
            return True

    # Verificar colunas
    for coluna in range(3):
        if tabuleiro[0][coluna] == jogador and tabuleiro[1][coluna] == jogador and tabuleiro[2][coluna] == jogador:
            return True

    # Verificar diagonais
    if tabuleiro[0][0] == jogador and tabuleiro[1][1] == jogador and tabuleiro[2][2] == jogador:
        return True
    if tabuleiro[0][2] == jogador and tabuleiro[1][1] == jogador and tabuleiro[2][0] == jogador:
        return True

    return False

def jogar():
    tabuleiro = criar_tabuleiro()
    jogador_atual = 'X'
    
    while True:
        imprimir_tabuleiro(tabuleiro)
        
        linha = int(input('Digite a linha (0-2): '))
        coluna = int(input('Digite a coluna (0-2): '))
        
        if tabuleiro[linha][coluna] != ' ':
            print('Posição ocupada. Tente novamente.')
            continue
        
        tabuleiro[linha][coluna] = jogador_atual
        
        if verificar_vitoria(tabuleiro, jogador_atual):
            imprimir_tabuleiro(tabuleiro)
            print(f'Jogador {jogador_atual} venceu!')
            break
        
        if jogador_atual == 'X':
            jogador_atual = 'O'
        else:
            jogador_atual = 'X'

jogar()
```
